```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.vector_ar.var_model import VAR
```

```python
factors = np.load('factors_iid.npy')
```

```python
plt.figure(figsize=(15,5))
plt.plot(factors);
```

## Estimating var process for factors

```python
model = VAR(endog=factors)
res = model.fit(maxlags = 2)
```

```python
plt.figure(figsize=(15,5))
plt.plot(res.fittedvalues, color='blue', linewidth=0.5);
plt.plot(factors, color='black', linewidth=0.5);
```

```python
plt.figure(figsize=(15,5))
plt.plot(res.resid, color='blue', linewidth=0.5);
```

```python
res.plot_sample_acorr();
```

```python
res.summary()
```
