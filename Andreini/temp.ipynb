## DIagonal Var(1) with nondiagonal Q

```python
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Construct the model
class DiagVar(sm.tsa.statespace.MLEModel):
    def __init__(self, endog):
        # Initialize the state space model
        self.k_states = endog.shape[1]
        super(DiagVar, self).__init__(endog, k_states=self.k_states, k_posdef=self.k_states,
                                  initialization='stationary')

        # Setup the fixed components of the state space representation
        self.ssm['design'] = np.eye(self.k_states)
        self.ssm['selection'] = np.eye(self.k_states)
        self.ssm['obs_cov'] = np.zeros((self.k_states, self.k_states))

        self.n_state_cov = int((self.k_states*(self.k_states+1)/2))


    # Describe how parameters enter the model
    def update(self, params, *args, **kwargs):
        params = super(DiagVar, self).update(params, *args, **kwargs)
        self.ssm['transition'] = np.diag(params[0:self.k_states])
        a = np.zeros((self.k_states,self.k_states))
        ind = np.tril_indices(self.k_states)
        a[ind] = params[self.k_states:]
        a = np.maximum(a,a.T)
        self.ssm['state_cov'] = a
    
    @property
    def param_names(self):
        return [f"T_{i},{i}" for i in range(1,self.k_states+1)] + [f"q_{i}" for i in range(1,self.n_state_cov+1)]

    # Specify start parameters and parameter names
    @property
    def start_params(self):
        cov = np.cov(self.endog,rowvar=False)
        cov_ind = np.tril_indices(self.k_states)
        cov_arr = cov[cov_ind].flatten()
        return [0] * self.k_states + list(cov_arr)
```

```python
def simulateVar(x0, delta=0, lamda=None, T=10000, mu=0, omega=None):
    assert delta.shape[0] == x0.shape[0] == lamda.shape[0] == x0.shape[0] == omega.shape[0] == mu.shape[0], 'Incorrect shapes'
    eigval, _ = np.linalg.eig(lamda)
    if not (np.abs(eigval)<=1.).all(): print(f"Warning: this system is unstable. Eigenvalues: {eigval}")
    res = [x0]
    for t in range(T):
        residuals = np.random.multivariate_normal(mu,omega)
        xt = delta + lamda @ res[-1] + residuals
        res.append(xt)
    return np.array(res)
```

```python
k = 3
lamda = np.diag(np.random.uniform(-1,1,size=k))
delta = np.zeros((k))
mu = np.zeros((k))
Q = np.random.uniform(0.1,0.9, size=(k,k))
Q = (Q @ Q.T)
x0 = np.random.multivariate_normal(np.zeros(k), Q)
y=simulateVar(x0,delta=delta,lamda=lamda,mu=mu,omega=Q)
plt.plot(y[0:100]);
print(lamda, '\n', Q)
```

```python
mod = DiagVar(y)
res = mod.fit(full_output=False, disp=False)
res.summary()
print(res.params)
sim = mod.simulate(res.params,nsimulations=100)
plt.plot(sim)
```
